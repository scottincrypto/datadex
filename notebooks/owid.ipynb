```python
from owid import catalog
import plotly.express as px

catalog.find(namespace="energy")
```

```python
energy_full = catalog.find(table="owid_energy", version="2023-07-10").iloc[0].load()
energy_full
```

```python
energy = energy_full[["population", "gdp", "primary_energy_consumption"]].reset_index()

energy = energy.dropna()


# energy['gdp_per_capita'] = energy['gdp'] / energy['population']
```

```python
energy["gdp_per_capita"] = energy["gdp"] / energy["population"]
energy["primary_energy_consumption_per_capita"] = (
    energy["primary_energy_consumption"] / energy["population"]
)
energy["country_string"] = energy["country"].astype(str)
energy = energy.sort_values(by=["country_string", "year"])
```

```python
# df = energy.filter('country_string == "Australia"')
df = energy
df = df.loc[
    df["country_string"].isin(
        [
            "Australia",
            "United States",
            "China",
            "India",
            "Japan",
            "United Kingdom",
            "Germany",
            "France",
            "Italy",
            "Canada",
            "Spain",
            "South Africa",
            "Brazil",
            "Russia",
        ]
    )
]

fig = px.line(
    df,
    x="primary_energy_consumption_per_capita",
    y="gdp_per_capita",
    color="country_string",
    hover_name="year",
)
fig.update_layout(width=1000, height=600)
fig.show()
```
